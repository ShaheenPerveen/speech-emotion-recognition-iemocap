```python
rmse1 = librosa.feature.rms(y1 + 0.0001)[0]
rmse2 = librosa.feature.rms(y2 + 0.0001)[0]
rmse3 = librosa.feature.rms(y3 + 0.0001)[0]
rmse4 = librosa.feature.rms(y4 + 0.0001)[0]
# plt.figure(figsize=(15,2))
plt.plot(rmse1, color='r')
plt.plot(rmse2, color='b')
plt.plot(rmse3, color='g')
plt.plot(rmse4, color='y')
plt.ylabel('RMSE')
```

```python
autocorr1 = librosa.core.autocorrelate(y1)
autocorr2 = librosa.core.autocorrelate(y2)
autocorr3 = librosa.core.autocorrelate(y3)
autocorr4 = librosa.core.autocorrelate(y4)
```

```python
plt.figure(figsize=(15,2))
plt.plot(autocorr1, color='r', alpha=0.5)
plt.plot(autocorr2, color='b', alpha=0.5)
plt.plot(autocorr3, color='g', alpha=0.5)
plt.plot(autocorr4, color='y', alpha=0.5)
plt.ylabel('Autocorrelations')
```
